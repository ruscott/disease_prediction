# Disease Predictor

This data set is taken from kaggle and shows symptoms and prognosis of these

```python
#Modules to use
import pandas as pd
import numpy as npy
```

```python
#Load and investigate data
train_df = pd.read_csv('Training.csv')
test_df = pd.read_csv('Testing.csv')

print(train_df.columns)
print(train_df.head(3))

train_df.drop(columns=['Unnamed: 133'], inplace=True)
```

```python
#remove uninformative variables
uninf_var = train_df.loc[:, train_df.nunique() < 2].columns
uninf_var = uninf_var.to_list()
print(uninf_var)
```

```python
#Function to clean data the same for train and test 
def df_clean(df):
    #Drop columns from both test and train using train uninformative variables
    df.drop(columns=uninf_var, inplace=True)
    print(df.head(2))

    #check to see if df has any null values and drop rows with null values
    null_values = df.columns[df.isnull().any()].to_list()
    print('\n Number of null values:', null_values, '\n')
    if len(null_values) > 0:
        df.dropna()
    else:
        next

    return df
```

```python
train_df = df_clean(train_df)
print(train_df.columns)
```

```python
test_df = df_clean(test_df)
print(test_df.head())
```

## Decision Trees

When assessing appropriate techniques, a decision tree was considered as an option. This is a classification process that will find the prognosis through making different decisions about the data. It follows a tree like structure moving from the top branch dependent on a set of yes and no rules as seen in the figure below. *** add figures ***. The rules of each branch are desired by considering the entropy and information gain of each rule set. The entropy measures the probabilistic chaos of the system (the randomness), so a smaller value is desired. The information gain is a statistical measures how much the choice made aids the decision, so a high value is desired. It is the difference in entropy of the system before and after the decision is made.

```python
#To make the decision tree the sklearn library was used.
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import mean_absolute_error
```

```python
#Set up train and test data
x_train = train_df.drop('prognosis', axis = 1)
y_train = train_df['prognosis']

x_test = test_df.drop('prognosis', axis = 1)
y_test = test_df['prognosis']
```

```python
tree = DecisionTreeClassifier()
```

```python
tree.fit(x_train, y_train)
```

```python
pred = tree.predict(x_test)
acc = tree.score(x_test, y_test)

print(acc)
```
